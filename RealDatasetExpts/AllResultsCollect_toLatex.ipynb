```python
import os
import sys
sys.path.append('../')

from Codes.model import *
from Codes.utils import *

import pickle

DATA_DIR = "../../Datasets"
RESULT_DIR = "../../DatasetsResult"
DATAINFO_DIR = "../../DatasetsInfo"
RESULTOUT_F_DIR = "../../ResultOutliers"
```

```python
datasets_list = ['Australian', 'Banknote',   'Codrna', 'Ecoli', 'Iris', 'Mammographic', 'Newthyroid', 'Skinsegment', 
                    'Wallrobot', 'Wine',  'Seeds','Vertebral','Winequality', 'Pendigits', 'blood', 'sepsis', 'balance', 'banana', 'census','shuttle'] 


d_list = ['Australian', 'banknote', 'codrna', 'ecoli', 'Iris', 'mammographic', 'Newthyroid', 'skinsegment', 
                    'wallrobot', 'Wine', 'seeds','vertebral','winequality','pendigits', 'blood', 'sepsis', 'balance', 'banana', 'census','shuttle'] 

```

```python

def extractData_Info(datasets_list,d_list ):

    df_data = pd.DataFrame(np.c_[datasets_list, np.zeros((len(datasets_list),3))] ,columns=(['Dataset', 'n','d' ,'k']))


    for i, data in enumerate(d_list):

        df_data.iloc[i,0] = datasets_list[i]

        with open(os.path.join(DATAINFO_DIR, str(data)+'.pkl'), 'rb') as inp:
            n = pickle.load(inp)
            d = pickle.load(inp)
            k = pickle.load(inp)

        df_data.loc[i,'n'] = n
        df_data.loc[i,'d'] = d
        df_data.loc[i,'k'] = k

        # df_data.loc[i,'Data source'] = str('\href{'+data_links[i]+'}{'+data_repo[i]+'}')
    
    return df_data
```

```python
df_data =  extractData_Info(datasets_list , d_list)
df_data = df_data.sort_values(by=['n'])
print(df_data.to_latex(index=False))
```

```python

def extractData_ci(datasets_list,d_list ):

    # from turtle import pu


    cols = ['kc-Gon', 'kc-GonAvg', 'kc-HS', 'kmed', 'kmed++', 'Opt-Cg']
    model = ['kcgon', 'kcgonavg', 'kchs', 'km', 'kmplus', 'mmopt']
    allcols = ['data', 'n','d', 'kc-Gon', 'kc-GonAvg', 'kc-HS', 'kmed', 'kmed++', 'Opt-Cg']


    df_ari = pd.DataFrame(np.c_[datasets_list, np.zeros((len(datasets_list),len(cols)+2))] ,columns=allcols)
    df_purity = pd.DataFrame(np.c_[datasets_list, np.zeros((len(datasets_list),len(cols)+2))] ,columns=allcols)
    df_nmi = pd.DataFrame(np.c_[datasets_list, np.zeros((len(datasets_list),len(cols)+2))] ,columns=allcols)
    df_epsVal = pd.DataFrame(np.c_[datasets_list, np.zeros((len(datasets_list),len(cols)+2))] ,columns=allcols)
    df_time = pd.DataFrame(np.c_[datasets_list, np.zeros((len(datasets_list),len(cols)+2))] ,columns=allcols)

    for i, data in enumerate(d_list):

        df_ari.iloc[i,0] = datasets_list[i]

        with open(os.path.join(DATAINFO_DIR, str(data)+'.pkl'), 'rb') as inp:
            n = pickle.load(inp)
            d = pickle.load(inp)
            k = pickle.load(inp)

        df_ari.loc[i,'n'] = n
        df_ari.loc[i,'d'] = d
        # df_ari.loc[i,'k'] = k
        
        df_purity.loc[i,'n'] = n
        df_purity.loc[i,'d'] = d        
        df_nmi.loc[i,'n'] = n
        df_nmi.loc[i,'d'] = d
        df_time.loc[i,'n'] = n
        df_time.loc[i,'d'] = d
        df_epsVal.loc[i,'n'] = n
        df_epsVal.loc[i,'d'] = d
        epsVal = np.zeros((6,1))
        ari = np.zeros((6,1))
        purity = np.zeros((6,1))
        time = np.zeros((6,1))



        with open(os.path.join(RESULT_DIR, str(data)+'.pkl'), 'rb') as inp:


            ari[0] =pickle.load(inp)
            purity[0]=pickle.load(inp)
            nmi_kcgon=pickle.load(inp)
            epsVal[0]=pickle.load(inp)
            centers_kcgon=pickle.load(inp)
            time[0]=pickle.load(inp)

            ari1=pickle.load(inp)
            purity1=pickle.load(inp)
            nmi_result_kcgonavg=pickle.load(inp)
            epsVal1=pickle.load(inp)
            centers_list_gonavg=pickle.load(inp)
            time[1]=pickle.load(inp)

            ari[2]=pickle.load(inp)
            purity[2]=pickle.load(inp)
            nmi_kchs=pickle.load(inp)
            epsVal[2]=pickle.load(inp)
            centers_kchs=pickle.load(inp)
            time[2]=pickle.load(inp)
            
            ari2=pickle.load(inp)
            purity2=pickle.load(inp)
            nmi_result_km=pickle.load(inp)
            epsVal2=pickle.load(inp)
            centers_list_km=pickle.load(inp)
            time[3]=pickle.load(inp)
            
            ari3=pickle.load(inp)
            purity3=pickle.load(inp)
            nmi_result_kmplus=pickle.load(inp)
            epsVal3=pickle.load(inp)
            centers_list_kmplus=pickle.load(inp)
            time[4]=pickle.load(inp)

            ari[5]=pickle.load(inp)
            purity[5]=pickle.load(inp)
            nmi_result_mmopt=pickle.load(inp)
            epsVal[5]=pickle.load(inp)
            centers_mmopt=pickle.load(inp)
            time[5]=pickle.load(inp)
            optgap_mmopt=pickle.load(inp)
            optVal_mmopt=pickle.load(inp)

            epsVal[1] = epsVal1[0]
            epsVal[3] = epsVal2[0]
            epsVal[4] = epsVal3[0]

            ari[1] = ari1[0]
            ari[3] = ari2[0]
            ari[4] = ari3[0]


            purity[1] = purity1[0]
            purity[3] = purity2[0]
            purity[4] = purity3[0]


        for j,mod in enumerate(model):

            df_epsVal.loc[i, cols[j]] = float('{:.1f}'.format(float(np.around(epsVal[j], 3))))
            df_ari.loc[i, cols[j]] = '{:.3f}'.format(float(np.around(ari[j], 3)))
            df_purity.loc[i, cols[j]] = '{:.3f}'.format(float(np.around(purity[j], 3)))
            df_time.loc[i, cols[j]] = '{:.2f}'.format(float(np.around(time[j], 3)))





    return df_ari, df_purity, df_nmi, df_epsVal, df_time
```

```python

df_ari, df_purity, df_nmi, df_epsVal, df_time =  extractData_ci(datasets_list , d_list)
df_epsVal = df_epsVal.drop(['kmed'], axis=1)
df_time = df_time.drop(['kmed'], axis=1)

df_epsVal = df_epsVal.sort_values(by=['n'])

df_time = df_time.sort_values(by=['n'])
```
